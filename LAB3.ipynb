```python
#3
import matplotlib.pyplot as plt
import csv

X = []
Y = []

with open('values1.txt', 'r') as datafile:
	plotting = csv.reader(datafile, delimiter=',')
	
	for ROWS in plotting:
		X.append(int(ROWS[0]))
		Y.append(int(ROWS[1]))

plt.plot(X, Y)
plt.title('Line Graph using CSV')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
#6
#i.
x=np.array(["Java","Python","PHP","Javascript","C#","C++"])
y=np.array([22.2,17.6,8.8,8,77,6.7])
plt.bar(x,y)
plt.title("Popularity")
plt.show()

#ii.
x=np.array(["Java","Python","PHP","Javascript","C#","C++"])
y=np.array([22.2,17.6,8.8,8,77,6.7])
plt.barh(x,y,color='red')
plt.title("Popularity- Horizontal")
plt.show()

#iii.
x=np.array(["Java","Python","PHP","Javascript","C#","C++"])
y=np.array([22.2,17.6,8.8,8,77,6.7])
plt.barh(x,y,color=['red','yellow','purple','blue','green','orange'])
plt.title("Popularity- Horizontal -Colors")
plt.show()
```

```python
#7
x = np.arange(5)
y1 = [22, 30, 35, 35, 26]
y2 = [25, 32, 30, 35, 29]
width = 0.3

plt.bar(x-0.3, y1, width, color='green')
plt.bar(x, y2, width, color='red')
plt.xticks(x, ['G1', 'G2', 'G3', 'G4', 'G5'])
plt.xlabel("Person")
plt.ylabel("Scores")
plt.legend(["Men", "Women"])
plt.title("Scores by person")
plt.show()
```
